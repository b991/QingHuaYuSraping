# Qing Hua Yu Data Processing and Visualizing Result

In this notebook, we processed the data that we collect from QingHuaYu website（https://allcp.net/）. We count how many poeple are online in 60-minute interval to gather what's the optimal time to pubish novel.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
QHYdf = pd.read_csv("./QingHuaYuResults.csv")
```

```python
Aug27th = QHYdf.loc[QHYdf['date']=="2023-09-15"]
```

Process date and time so that combined_datetime is a datetime object

```python
QHYdf['combined_datetime'] = QHYdf['date'] + ' ' +QHYdf['time']
QHYdf['combined_datetime']=pd.to_datetime(QHYdf['combined_datetime'])
```

Convert UTC to Beijing Standard Time. Seperate converted date from converted time. 

It's important to make converted_time into int, otherwise the plot won't come out correctly. 

```python
QHYdf['converted_date'] = QHYdf['combined_datetime'].dt.tz_localize('UTC').dt.tz_convert('Asia/Shanghai').dt.strftime('%Y-%m-%d')
QHYdf['converted_time'] = QHYdf['combined_datetime'].dt.tz_localize('UTC').dt.tz_convert('Asia/Shanghai').dt.strftime('%H').astype(int)
```

# Don't touch the dataframe above

```python
grouped_df = QHYdf.groupby('date').size().reset_index(name='count')
```

Create a list, add table of each day in the list

```python
dataList = []
for i in range(0,len(grouped_df)):
    obj = QHYdf.loc[QHYdf['date']== grouped_df.loc[i,"date"]].sort_values('converted_time').reset_index()
    dataList.append(obj)
```

Plot values

```python
for table in dataList:
    l = table.loc[0,"converted_date"]
    plt.plot(table["converted_time"],table["count"], label=l)
plt.legend(loc="upper left")
plt.xlabel('Hour of the Day, China Standard Time')
plt.ylabel('People Count')
```
